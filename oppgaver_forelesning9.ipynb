```python
# Oppgave 9.1 Lese lottotall fra fil
# Filen inneholder de 7 lottotallene som er trukket hver uke de siste 52 ukene. Programmet
# skal s ̊a finne ut hvor ofte tallet 2 er har vært med i vinnerrekka disse 52 ukene. Hint: nr2 =
# np.count nonzero(lotto == 2).

import numpy as np
url = "https://techteach.no/python/files/lotto.txt"

data = np.loadtxt(url)

nr2 = np.count_nonzero(data == 2)

print("Tallet 2 har vært med", nr2, "ganger")
```

```python
# Oppgave 9.2 Lese klimadata fra fil
# I kolonne 1 st ̊ar  ̊arstall og i kolonne 2 finner du middeltemperaturen for september et sted i
# Porsgrunn. Bruk data fra klima.txt til  ̊a finne maks- og min-temperaturen i dette
# datasettet. Videre skal du skrive ut  ̊aret disse maks- og min-temperaturene ble m ̊alt.

import numpy as np
url = "https://techteach.no/python/files/klima.txt"

data = np.loadtxt(url)

aar = data[:, 0] 
temp = data[:, 1] 

maks_temp = max(temp)
min_temp = min(temp)

maks_aar = aar[np.argmax(temp)]
min_aar = aar[np.argmin(temp)]

print( "Maks temp:", maks_temp)
print( " -> målt i år", int(maks_aar)) 
print( "Min temp:", min_temp)
print (" -> målt i år", int(min_aar))



```

```python
# Beregn og plott gjennomsnittstemperaturen fra 1961 til 1990 sammen med
# temperaturverdiene i klima.txt.

import numpy as np
import matplotlib.pyplot as plt

# Last ned data
url = "https://techteach.no/python/files/klima.txt"
data = np.loadtxt(url)

# Del opp data i årstall og temperaturer
aar = data[:, 0]  # Første kolonne: årstall
temp = data[:, 1]  # Andre kolonne: temperaturer

# Filtrer ut data for perioden 1961-1990
maske = (aar >= 1961) & (aar <= 1990)  # Boolsk maske for år i intervallet
aar_filtrert = aar[maske]  # Filtrerte årstall
temp_filtrert = temp[maske]  # Filtrerte temperaturer

# Beregn gjennomsnittstemperatur for perioden
gjennomsnitt_temp = np.mean(temp_filtrert)

# Plott temperaturdata for perioden 1961-1990
plt.plot(aar_filtrert, temp_filtrert, label="Temperatur (1961-1990)", color="black")

# Plott gjennomsnittstemperaturen som en horisontal linje
plt.axhline(y=gjennomsnitt_temp, color='red', linestyle='--', label=f"Gj.snitt: {gjennomsnitt_temp:.2f}°C")

# Sett opp aksene
plt.xlabel("År")
plt.ylabel("Temperatur (°C)")
plt.title("Temperaturdata for Porsgrunn (1961-1990)")
plt.legend()
plt.grid()

# Vis plottet
plt.show()

# Skriv ut gjennomsnittstemperaturen
print(f"Gjennomsnittstemperatur for perioden 1961-1990: {gjennomsnitt_temp:.2f}°C")



```
