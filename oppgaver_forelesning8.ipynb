```python
# Oppgave 8.6 Linspace
# Bruk numpys linspace-funksjon til  ̊a generere 20 x-koordinater p ̊a intervallet [0, π]. Opprett
# funksjonen f (x) = sin(x). Lag s ̊a en for-løkke som fortløpende skriver x-verdiene og de
# tilhørende funksjonsverdiene til skjerm med passende tekst.

import numpy as np

def fun(x):
    return np.sin(x)

x = np.linspace(0, np.pi, 20)

for i in x:
    print("For x =", round(i, 2), "er f=", round(fun(i), 3))
```

```python
# Oppgave 8.12 While-løkke
# Lag et program med en while-løkke som summerer partallene 0 + 2 + 4 + ... Løkken
# avsluttes n ̊ar summen ≥ 2250. Hvilket partall gjør at summen overstiger 2250?

sum = 0  # Startverdien for summen
num = 0  # Første partall

while sum < 2250:
    sum += num  # Legg til partallet i summen
    num += 2  # Øk til neste partall

# Når løkken stopper, har det siste partallet fått summen til å overstige 2250
print("Summen overstiger 2250 ved partallet:", num - 2)
```

```python
# Oppgave 8.13 Summering med while-løkke
# Opprettt en array x som inneholder de 20 heltallene 3, 4, 5, 6,..., 22, med følgende kode: x
# = np.linspace(3, 22, 20). Bruk en while-løkke til  ̊a summere elementene 3+4+5+... inntil
# summen blir større eller lik 150. Antallet elementer som inngikk i summen skrives til skjerm.

import numpy as np

x = np.linspace(3, 22, 20)  # Lager en array med 20 verdier fra 3 til 22
i = 0
total_sum = 0  # Bruk et annet navn enn "sum" for å unngå konflikt med innebygd sum()-funksjon

while (i < len(x) and total_sum < 150):  # Stopper når total_sum >= 150 eller vi når slutten av x
    total_sum += x[i]
    i += 1

print(f"Summen ble {total_sum}, og det krevde {i} elementer.")
    
```

```python
# Oppgave 8.14 Halvering med while-løkke
# Gitt at x = 10. Halverer man x blir svaret 5. Halverer man det svaret en gang til f ̊ar man
# resultat 2,5. Bruk en while-løkke til suksessivt  ̊a halvere x inntil svaret blir mindre eller lik
# 0,04. Antall halveringer som m ̊atte til for  ̊a oppn ̊a dette skrives til skjerm.

x = 10
antall = 0

while (x > 0.04):
    antall += 1
    x = x/2
print(x)
print(antall)
    
```

```python
# Oppgave 8.15 Dobling med while-løkke
# Gitt at x = 0,01. Dobler man x blir svaret 0,02. Dobler man det svaret en gang til f ̊ar man
# resultat 0,04. Bruk en while-løkke til suksessivt  ̊a doble x inntil svaret blir større enn 2000.
# Antall doblinger som m ̊atte til for  ̊a oppn ̊a dette skrives til skjerm.

x = 0.01
antall = 0

while (x <= 2000):
    antall += 1
    x = x*2

print(x)
print(antall)
```

```python
# Oppgave 8.18 While-indeks
# Opprett en array x som inneholder de 20 heltallene 3, 4, 5, 6,..., 22, se oppg 8.13. Bruk en
# while-løkke til  ̊a finne den minste indeksen i slik at np.sum(x[0:i]) > np.sum([i+1,:]).
# Indeksen i skal skrives til skjerm.
import numpy as np

# Oppretter arrayen x med verdiene 3 til 22
x = np.arange(3, 23)

# Initialiserer indeksen i
i = 1
n = len(x)

# Bruker en while-løkke til å finne den minste i som oppfyller betingelsen
while i < n:
    if np.sum(x[:i]) > np.sum(x[i+1:]):  # Sjekker betingelsen
        print(f"Den minste indeksen i er: {i}")
        break  # Stopper løkken når betingelsen er oppfylt
    i += 1
    
```
